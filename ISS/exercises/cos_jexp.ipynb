```python
import numpy as np
import matplotlib.pyplot as plt
# >>>>>>>>>>>>>>>>> #cosine_basic <<<<<<<<<<<<<<<<<<<<<<<<<
t = np.arange(0, 2*np.pi, 0.05)
x = np.cos(t)
plt.figure(figsize=(10,5))
plt.plot(t,x) 
plt.gca().set_xticks([0, np.pi/2, np.pi, 3*np.pi/2, 2*np.pi])
plt.gca().set_xticklabels(['0', '$\pi/2$', '$\pi$', '$3\pi/2$', '$2\pi$'])
plt.gca().set_xlabel('$t$')
plt.gca().set_ylabel('$x(t)$')
plt.gca().grid()
##plt.savefig('FIG/cos_basic.pdf')
```

```python
# >>>>>>>>>>>>>>>>> #cosine_full <<<<<<<<<<<<<<<<<<<<<<<<<
t = np.arange(-5,15,0.05)
x1 = np.cos(2*np.pi*0.5*t)
x2 = 10*np.cos(2*np.pi*0.5*t)
x3 = 10*np.cos(2*np.pi*0.5*t+0.5*np.pi)     #<<< try to chaneg the phase here ! 
_, ax = plt.subplots(3,1, figsize=(10,5))
ax[0].plot(t,x1)
ax[0].set_title('$\cos(2 \pi 0.5 t)$')
ax[0].set_xticklabels([])
ax[0].grid()
ax[1].plot(t,x2)
ax[1].set_title('$10\cos(2 \pi 0.5 t)$')
ax[1].set_xticklabels([])
ax[1].grid()
ax[2].plot(t,x3)
ax[2].set_title('$10\cos(2 \pi 0.5 t+ \pi/2)$')
ax[2].grid()
##plt.savefig('FIG/cos.pdf')
```

```python
# >>>>>>>>>>>>>>>>> #cosine_disc <<<<<<<<<<<<<<<<<<<<<<<<<
N = 50 
n = np.arange(0,N)
x1 = np.cos(2*np.pi/N*n)
x2 = 10*np.cos(2*np.pi/N*n + np.pi/2)
x3 = np.cos(5*2*np.pi/N*n)
_, ax = plt.subplots(3,1, figsize=(10,5))
ax[0].stem(n,x1,basefmt=" ", use_line_collection="true")  # don't want the baseline  ...
ax[0].set_title('$\cos(2 \pi n / N)$')
ax[0].set_xticklabels([])
ax[0].grid()
ax[1].stem(n,x2,basefmt=" ", use_line_collection="true")  # dtto ...
ax[1].set_title('$10\cos(2 \pi n / N + \pi  /2)$')
ax[1].set_xticklabels([])
ax[1].grid()
ax[2].stem(n,x3,basefmt=" ", use_line_collection="true")  # dtto ...
ax[2].set_title('$\cos(2 \pi 5 n / N)$')
ax[2].grid()
##plt.savefig('FIG/diskcos.pdf')
```

```python
# >>>>>>>>>>>>>>>>> #complex_exp <<<<<<<<<<<<<<<<<<<<<<<<<
from mpl_toolkits import mplot3d
import mpl_toolkits.mplot3d.art3d as art3d

# a "wire barrel" to see better what happens ... 
def complex_cylinder(t1, t2, r, fig):
    '''kresli sedy drateny valec ve 3d tak, aby byla dobre videt komplexni exponenciala. 
    t1 je pocatecni cas
    t2 je koncovy cas 
    r je polomer
    fig je obrazek'''
    phi = np.linspace(0,2 * np.pi,100)
    t = np.linspace(t1,t2,9)
    kolo = r * np.exp(1j * phi)
    for tt in t: 
        fig.plot3D(tt * np.ones(phi.shape), np.real(kolo), np.imag(kolo),color='k',alpha=0.1)
    phi = np.linspace(0,2 * np.pi,32)
    kolo = r * np.exp(1j * phi)
    for z in kolo:        
        line=art3d.Line3D(*zip((t1,np.real(z),np.imag(z)), (t2,np.real(z),np.imag(z))), markevery=(1, 1), color='k',alpha=0.1)
        ax.add_line(line)   
        
t = np.arange(0,2*np.pi,0.05)
x = np.exp(1j*t)
plt.figure(figsize=(10,5))
ax = plt.axes(projection="3d")
ax.plot3D(t, np.real(x), np.imag(x))
ax.set_xlabel('$t$')
ax.set_ylabel('$real(x(t))$')
ax.set_yticks([-1, 0, 1])
ax.set_zticks([-1, 0, 1])
ax.set_zlabel('$imag(x(t))$')
complex_cylinder(0,2*np.pi,1,ax)

##plt.savefig('FIG/kexp.pdf')
```

```python
# >>>>>>>>>>>>>>>>> #complex_exp_minus <<<<<<<<<<<<<<<<<<<<<<<<<
t = np.arange(0,2*np.pi,0.05)
x = np.exp(1j*t)
plt.figure(figsize=(10,5))
ax = plt.axes(projection="3d")
ax.plot3D(t, np.real(x), np.imag(x),color='b',label='$e^{jt}$')
ax.plot3D(t, np.real(x), -np.imag(x),color='r',label='$e^{-jt}$')
ax.legend()
ax.set_xlabel('$t$')
ax.set_ylabel('$real(x(t))$')
ax.set_yticks([-1, 0, 1])
ax.set_zticks([-1, 0, 1])
ax.set_zlabel('$imag(x(t))$')
complex_cylinder(0,2*np.pi,1,ax)

#plt.savefig('FIG/exp_plus_minus.pdf')
```

```python
# >>>>>>>>>>>>>>>>> #complex_exp_magnitude <<<<<<<<<<<<<<<<<<<<<<<<<
t = np.arange(0,2*np.pi,0.05)
x = np.exp(1j*t)
bigx = 2*np.exp(1j*t)
smallx = 2/3*np.exp(1j*t)
plt.figure(figsize=(10,5))
ax = plt.axes(projection="3d")
ax.plot3D(t, np.real(x), np.imag(x),color='b',label='$e^{jt}$')
ax.plot3D(t, np.real(bigx), np.imag(bigx),color='r',label='$2e^{jt}$')
ax.plot3D(t, np.real(smallx), np.imag(smallx),color='g',label='$2/3e^{jt}$')
ax.legend()
ax.set_xlabel('$t$')
ax.set_ylabel('$real(x(t))$')
ax.set_yticks([-2, 0, 2])
ax.set_zticks([-2, 0, 2])
ax.set_zlabel('$imag(x(t))$')
complex_cylinder(0,2*np.pi,1,ax)
complex_cylinder(0,2*np.pi,2,ax)
complex_cylinder(0,2*np.pi,2/3,ax)

##plt.savefig('FIG/exp_zmena_modulu.pdf')
```

```python
# >>>>>>>>>>>>>>>>> #complex_exp_phase <<<<<<<<<<<<<<<<<<<<<<<<<
t = np.arange(0,2*np.pi,0.05)
x = np.exp(1j*t)
x1 = np.exp(1j*np.pi/4)*np.exp(1j*t)
x2 = np.exp(-1j*2*np.pi/3)*np.exp(1j*t)
plt.figure(figsize=(10,5))
ax = plt.axes(projection="3d")
ax.plot3D(t, np.real(x), np.imag(x),color='b',label='$e^{jt}$')
ax.plot3D(t, np.real(x1), np.imag(x1),color='r',label='$e^{j\pi/4}e^{jt}$')
ax.plot3D(t, np.real(x2), np.imag(x2),color='g',label='$e^{-j2\pi/3}e^{jt}$')
ax.set_xlabel('$t$')
ax.set_ylabel('$real(x(t))$')
ax.set_yticks([-1, 0, 1])
ax.set_zticks([-1, 0, 1])
complex_cylinder(0,2*np.pi,1,ax)
ax.set_zlabel('$imag(x(t))$')
ax.legend()
##plt.savefig('FIG/exp_predtoceni.pdf')
```

```python
# >>>>>>>>>>>>>>>>> #complex_exp_magnitude_phase <<<<<<<<<<<<<<<<<<<<<<<<<
#mode, $1.5 e^{j\frac{\pi}{4}} e^{jx}$ erven a
#$\frac{1}{2}e^{-j\frac{2\pi}{3}} e^{jx}$ zelen. 
t = np.arange(0,2*np.pi,0.05)
x = np.exp(1j*t)
x1 = 1.5*np.exp(1j*np.pi/4)*np.exp(1j*t)
x2 = 0.5*np.exp(-1j*2*np.pi/3)*np.exp(1j*t)
plt.figure(figsize=(10,5))
ax = plt.axes(projection="3d")
ax.plot3D(t, np.real(x), np.imag(x),color='b',label='$e^{jt}$')
ax.plot3D(t, np.real(x1), np.imag(x1),color='r',label='$1.5e^{j\pi/4}e^{jt}$')
ax.plot3D(t, np.real(x2), np.imag(x2),color='g',label='$0.5e^{-j2\pi/3}e^{jt}$')
#ax.plot3D(np.zeros(t.shape), np.real(x), np.imag(x),color='k')
ax.legend()
ax.set_xlabel('$t$')
ax.set_ylabel('$real(x(t))$')
ax.set_yticks([-1, 0, 1])
ax.set_zticks([-1, 0, 1])
ax.set_zlabel('$imag(x(t))$')
complex_cylinder(0,2*np.pi,1,ax)
complex_cylinder(0,2*np.pi,1.5,ax)
complex_cylinder(0,2*np.pi,0.5,ax)
##plt.savefig('FIG/exp_modul_i_predtoceni.pdf')
```

```python
# >>>>>>>>>>>>>>>>> #complex_exponential_ang_freq  <<<<<<<<<<<<<<<<<<<<<<<<<
om1 = 2000*np.pi # kruh frekvence
T1 = 2*np.pi / om1 # perioda 
t = np.linspace(0,T1,num=200)
x = np.exp(1j*om1*t)
xm = np.exp(-1j*om1*t)
plt.figure(figsize=(10,5))
ax = plt.axes(projection="3d")
ax.plot3D(t, np.real(x), np.imag(x),color='b',label='$e^{j\omega_1 t}$')
ax.plot3D(t, np.real(xm), np.imag(xm),color='r',label='$e^{-j\omega_1 t}$')
ax.legend()
ax.set_xlabel('$t$')
ax.set_ylabel('$real(x(t))$')
ax.set_yticks([-1, 0, 1])
ax.set_zticks([-1, 0, 1])
ax.set_zlabel('$imag(x(t))$')
complex_cylinder(0,T1,1,ax)
##plt.savefig('FIG/exp_perioda_plus_minus.pdf')
```

```python
# >>>>>>>>>>>>>>>>> #cos_decomposition  <<<<<<<<<<<<<<<<<<<<<<<<<
plt.figure(figsize=(8,5))
ax = plt.axes(projection="3d")
t = np.linspace(0,T1,num=200)

xplus = 5/2*np.exp(1j*om1*t)
xminus = 5/2*np.exp(-1j*om1*t)
ax.plot3D(t, np.real(xplus), np.imag(xplus))
ax.plot3D(t, np.real(xminus), np.imag(xminus))
complex_cylinder(0,T1,2.5,ax)

x = 5*np.cos(om1*t)
aux1 = np.pad(t, (1,1), 'constant', constant_values=(0,np.max(t))) 
aux2 = np.pad(x, (1,1), 'constant', constant_values=(0,0))
aux3 = np.pad(np.zeros(t.shape), (1,1), 'constant', constant_values=(0,0))
verts = [list(zip(aux1,aux2,aux3))]
cos2d = mplot3d.art3d.Poly3DCollection(verts, alpha=0.5, linewidth=1)
cos2d.set_facecolor((1, 0, 0, 0.2))   # 0.2 is alpha
ax.add_collection3d(cos2d)

ax.set_xlabel('$t$')
ax.set_ylabel('$real(x(t))$')
ax.set_zlim3d(-5,5)
ax.set_ylim3d(-5,5)

ax.set_yticks([-5, 0, 5])
ax.set_zticks([-5, 0, 5])
#ax.set_xticks([0, 0.01, 0.02])
ax.set_zlabel('$imag(x(t))$')
ax.view_init(elev=45, azim=-40)

##plt.savefig('FIG/cosomegat.pdf')
```

```python
# >>>>>>>>>>>>>>>>> #cos_decomposition_full  <<<<<<<<<<<<<<<<<<<<<<<<<
om1 = 100*np.pi # kruh frekvence
T1 = 2*np.pi / om1 # perioda 
t = np.linspace(0,T1,num=200)
plt.figure(figsize=(8,5))

ax = plt.axes(projection="3d")
t = np.linspace(0,T1,num=200)
xplus = 5/2*np.exp(-1j*np.pi/4)*np.exp(1j*om1*t)
xminus = 5/2*np.exp(1j*np.pi/4)*np.exp(-1j*om1*t)
ax.plot3D(t, np.real(xplus), np.imag(xplus))
ax.plot3D(t, np.real(xminus), np.imag(xminus))
complex_cylinder(0,T1,2.5,ax)

x = 5*np.cos(om1*t-np.pi/4)
aux1 = np.pad(t, (1,1), 'constant', constant_values=(0,np.max(t))) 
aux2 = np.pad(x, (1,1), 'constant', constant_values=(0,0))
aux3 = np.pad(np.zeros(t.shape), (1,1), 'constant', constant_values=(0,0))
verts = [list(zip(aux1,aux2,aux3))]
cos2d = mplot3d.art3d.Poly3DCollection(verts, alpha=0.5, linewidth=1)
cos2d.set_facecolor((1, 0, 0, 0.2))   # 0.2 is alpha
ax.add_collection3d(cos2d)

ax.set_xlabel('$t$')
ax.set_ylabel('$real(x(t))$')
ax.set_zlim3d(-5,5)
ax.set_ylim3d(-5,5)

ax.set_xticks([0, 0.01, 0.02])
ax.set_yticks([-5, 0, 5])
ax.set_zticks([-5, 0, 5])
ax.set_zlabel('$imag(x(t))$')
ax.view_init(elev=45, azim=-40)

##plt.savefig('FIG/cosphi.pdf')
```

```python
# >>>>>>>>>>>>>>>>> #discrete_cos_decomposition  <<<<<<<<<<<<<<<<<<<<<<<<< 
n = np.arange(0,32)
x = 5*np.cos(2*np.pi/32 * n - np.pi/3)
xplus = 2.5*np.exp(-1j*np.pi/3)*np.exp(1j * 2*np.pi/32 * n)
xminus = 2.5*np.exp(1j*np.pi/3)*np.exp(1j * 2*np.pi/32 * n)

plt.figure(figsize=(10,6))
ax = plt.axes(projection="3d")
for xi, yi, zi in zip(n, np.real(xplus), np.imag(xplus)):        
    line=art3d.Line3D(*zip((xi, 0, 0), (xi, yi, zi)), marker='o', markevery=(1, 1),color='b')
    ax.add_line(line)
for xi, yi, zi in zip(n, np.real(xminus), np.imag(xminus)):        
    line=art3d.Line3D(*zip((xi, 0, 0), (xi, yi, zi)), marker='o', markevery=(1, 1),color='r')
    ax.add_line(line)
for xi, yi, zi in zip(n, x, np.zeros(n.shape)):        
    line=art3d.Line3D(*zip((xi, 0, 0), (xi, yi, zi)), marker='o', markevery=(1, 1),color='k')
    ax.add_line(line)
complex_cylinder(0,32,2.5,ax)
   
ax.set_xlim3d(0, 32)
ax.set_ylim3d(-5, 5)
ax.set_zlim3d(-5, 5)
ax.set_yticks([-5, 0, 5])
ax.set_zticks([-5, 0, 5])
ax.set_xlabel('$n$')
ax.set_ylabel('$real(x[n])$')
ax.set_zlabel('$imag(x[n])$')

ax.view_init(elev=45, azim=-40)
##plt.savefig('FIG/dcos_rozklad.pdf')
```
